```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = []
y = []
for i in range(-100,100,1):
    x.append(i*0.1)
    y.append(1/(1+np.exp(-i*0.1)))
```

```python
plt.plot(x,y,label = 'sigmoid')
plt.xlabel('input')
plt.ylabel('sigmoid')
plt.legend(loc='upper right')
```
