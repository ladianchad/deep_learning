```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([(0,0),(1,0)
     ,(0,1),(1,1)])
Y = np.array([0,1,1,0])

plt.plot(X,Y,'ro')
plt.show()
```

```python
class logistic_regression():
    def __init__(self,dim = 2,method = 'standard_normal'):
        if method == 'rand':
            self.w = np.random.rand(dim)
            self.b = np.random.rand()
        elif method == 'randn':
            self.w = np.random.randn(dim)
            self.b = np.random.randn()
        elif method == 'standard_normal':
            self.w = np.random.standard_normal(dim)
            self.b = np.random.standard_normal()
        else:
            self.w = np.random.standard_normal(dim)
            self.b = np.random.standard_normal()
    
    def sigmoid(self,z):
        return 1/(1+np.exp(-z))
    
    def predict(self,X):
        z = X@self.w + self.b
        return self.sigmoid(z)
    
def train(X,Y,model,lr=0.1):
    dw = np.zeros((self.w.shape[0]))
    db = 0.0
    m = len(X)
    cost = 0.0
    for x,y in zip(X,Y):
        pred = model.predict(x)
        cost -= np.log(pred) if y else np.log(1-pred)
        dw += (pred-y)*x
        db +=(pred -y)
    cost /= m
    model.w -= lr*dw/m
    model.b -= lr*db/m
    return cost


def show_graph(x_data = None,y_data=None,line_type='-'):
    if y_data is not None:
        if x_data is not None:
            plt.plot(x_data,y_data,line_type)
        else :
            plt.plot(y_data,line_type)
        plt.show()
    else:
        print ("no input data!!")
    
```

```python
logi = logistic_regression(method='rand')
```

```python
costs = []
w_params = []
b_params = []
for epoch in range(10000):
    cost = train(X,Y,logi,0.1)
    costs.append(cost)
    w_params.append(logi.w)
    b_params.append(logi.b)
```

```python
plt.plot(costs)
plt.xlabel('epoch')
plt.ylabel('cost')
plt.show()

plt.plot(w_params)
plt.xlabel('epoch')
plt.ylabel('w_params')
plt.show()

plt.plot(b_params)
plt.xlabel('epoch')
plt.ylabel('b_params')
plt.show()
```
